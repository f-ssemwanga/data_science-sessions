```python
#loading data
####Essential manual checks####
#missing values
#convert categorical values into numeric
#cehck non type values and remove them
#remove empty rows

data = pd.read_csv('/content/drive/MyDrive/HousePriceData/AdvertData.csv')
data
```

```python
##Brodatder feature
## if the number of rows do not match it means there are missing values
data.info()
```

```python
#Perform data preprocessing steps 
##Replacing the missing values by median of the Red_code column and Year of Registration

#go to columns with missing data and fill missing with media  - for numerical columns
data['reg_code'] = data['reg_code'].fillna(data['reg_code'].median())
data['year_of_registration'] = data['year_of_registration'].fillna(data['year_of_registration'].median())
data.info()
```

```python
###remove rows having numeric values in categorical columns - perform this step before Label encoder
#the tilda refers to just this column
dfnew = data[~data['standard_model'].str.contains('\d+')]
dfnew
```

```python
##remove empty row
df_final = dfnew.dropna(how="any", axis =0)
df_final
```

```python
##Handling erraoneous numeric values  i.e. take values greater than 0
df_final = df_final[(df_final['mileage'] > 0) & (df_final['price'] > 0) & (df_final['year_of_registration'] > 0)]
df_final
```

```python
###preprocessing for categorical values by converting them using OneHotEncoder - label encoder
le = LabelEncoder()
data['standard_colour'] = le.fit_transform(data['standard_colour'])
data['fuel_type'] = le.fit_transform(data['fuel_type'])
data['body_type'] = le.fit_transform(data['body_type'])
data['crossover_car_and_van'] = le.fit_transform(data['crossover_car_and_van'])
data['standard_model'] = le.fit_transform(data['standard_model'])
data.info()

```
