```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, LabelEncoder
```

```python
#loading data
####Essential manual checks####
#missing values
#convert categorical values into numeric
#cehck non type values and remove them
#remove empty rows

data = pd.read_csv('/content/drive/MyDrive/HousePriceData/AdvertData.csv')
data
```

```python
##Broader feature
## if the number of rows do not match it means there are missing values
data.info()
```

```python
#Perform data preprocessing steps
##Replacing the missing values by median of the Red_code column and Year of Registration

#go to columns with missing data and fill missing with media  - for numerical columns
data['reg_code'] = data['reg_code'].fillna(data['reg_code'].median())
data['year_of_registration'] = data['year_of_registration'].fillna(data['year_of_registration'].median())
data.info()
```

```python
###remove rows having numeric values in categorical columns - perform this step before Label encoder
#the tilda refers to just this column
dfnew = data[~data['standard_model'].str.contains('\d+')]
dfnew
```

```python
##remove empty row
df_final = dfnew.dropna(how="any", axis =0)
df_final
```

```python
##Handling erraoneous numeric values  i.e. take values greater than 0
df_final = df_final[(df_final['mileage'] > 0) & (df_final['price'] > 0) & (df_final['year_of_registration'] > 0)]
df_final
```

```python
#handling outliers - as they degrade the model's performace####
df_final['mileage'] = pd.to_numeric(df_final['mileage'], errors='coerce') #fills outliers with nearest value
q1 = np.percentile(df_final['mileage'], 25) # calculate the 25% of percentile
q3 = np.percentile(df_final['mileage'], 75) # calculate the 75 Percentetile
iqr = q3-q1 # calculate the interquartile range - used to identify outliers
threshold = 1.5*iqr # identify the outliers

#remove outliners
df_final = df_final[(df_final['mileage'] >= q1 - threshold) & (df_final['mileage'] <= q3 + threshold)]
print(f"Array withtout outliers:\n, {df_final}")
```

```python
###preprocessing for categorical values by converting them using OneHotEncoder - label encoder
##Label##
le = LabelEncoder()
import warnings
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    df_final.loc[:, 'standard_colour'] = le.fit_transform(df_final['standard_colour'])
    df_final.loc[:, 'standard_colour'] = le.fit_transform(df_final['standard_colour'])
    df_final.loc[:,'standard_make'] = le.fit_transform(df_final['standard_make'])
    df_final.loc[:,'standard_model'] = le.fit_transform(df_final['standard_model'])
    df_final.loc[:,'body_type'] = le.fit_transform(df_final['body_type'])
    df_final.loc[:,'crossover_car_and_van'] = le.fit_transform(df_final['crossover_car_and_van'])
    df_final.loc[:,'fuel_type'] = le.fit_transform(df_final['fuel_type'])
    df_final.loc[:,'vehicle_condition'] = le.fit_transform(df_final['vehicle_condition'])
df_final

```

```python
### indentify predictors and Target Variables also known as Independent and Dependent variables respectively from the dataset
# price is the good targer variable
#mileage

columns_to_drop = ['price', 'crossover_car_and_van', 'vehicle_condition']

x =df_final.drop(columns_to_drop, axis=1) ###preductirs /Indepedent variable features

y=df_final['price'] ##outcome variable

print("Independent / Predictors:", x)
print("Dependent / Outcome:", y)
```

```python
#split data in training and testing sets ###
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=42)
print("Training Independent Data", x_train)
print("Testing Independent Data", x_test)
print("Training dependent Data", y_train)
print("Testing Depedent Data", y_test)
```

```python
###IMplement linear regression on the data
###Linear regression is best for numberical data

## for categorical data use logistic regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score #tunning parameters for checking model accuracy

lr_classifier = LinearRegression()
lr_classifier.fit(x_train, y_train) # training the model


###predicting calues ##

lr_y_predict = lr_classifier.predict(x_test)
print(lr_y_predict)


##Evaluastion metrics###
mse = mean_squared_error(y_test, lr_y_predict) ##mean square error - how close the regression line to the data point
rmse = np.sqrt(mse) ## sqaureroot of mse
mae = mean_absolute_error(y_test, lr_y_predict) # mean absolute error
r2 = r2_score(y_test, lr_y_predict) #RSquare measures

print("Mean Squaed Error:", mse)
print("Root  Mean Squaed Error:", rmse)
print("Mean absolute Error:", mae)
print("Rsquared:", r2)


```
